# Pandas Visualization

Pandas menyediakan fungsi bawaan untuk membuat visualisasi data dengan cepat dan mudah. Berikut adalah berbagai jenis plot yang dapat dibuat dengan Pandas dan cara melakukan kustomisasi.

---

## 1. **Line Plot**

Line plot digunakan untuk memvisualisasikan data kontinu, seperti tren atau perubahan nilai terhadap waktu.

### Contoh:
```python
import pandas as pd

# Membuat DataFrame
data = {'Year': [2018, 2019, 2020, 2021, 2022],
        'Sales': [50, 60, 70, 85, 90]}
df = pd.DataFrame(data)

# Membuat Line Plot
df.plot(x='Year', y='Sales', kind='line', title='Line Plot: Sales Over Years')
```

---

````python
import pandas as pd

# Membuat DataFrame
data = {'Year': [2018, 2019, 2020, 2021, 2022],
        'Sales': [50, 60, 70, 85, 90]}
df = pd.DataFrame(data)
df.plot(x='Year', y='Sales', kind='line', title='Line Plot: Sales Over Years', marker='o', color='red', grid= True, linestyle='--')
````

## 2. Kustomisasi dengan Matplotlib

Setelah membuat plot dengan Pandas, kita bisa memanfaatkan fungsi-fungsi Matplotlib untuk mempercantik grafik.

Beberapa fungsi Matplotlib yang sering digunakan:
- **Judul**: `plt.title()`
- **Label Sumbu**: `plt.xlabel()`, `plt.ylabel()`
- **Grid**: `plt.grid()`
- **Legenda**: `plt.legend()`

### Contoh Kustomisasi:
```python
# Line plot dengan kustomisasi
df.plot(x='Bulan', y='Penjualan', kind='line', color='green')
plt.title('Penjualan Bulanan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.grid(True)
plt.show()
```




````python
df.plot(x='Year', y='Sales', kind='line', marker='o', color='red', grid= True, linestyle='--')
plt.xlabel('Year', fontsize= 14, color='blue')
plt.title('Line Plot: Sales Over Years',fontsize = 16)
````

## 2. **Bar Plot**

Bar plot cocok digunakan untuk data kategori atau membandingkan nilai antar kategori.

### Contoh:
```python
# Membuat Bar Plot
df.plot(x='Year', y='Sales', kind='bar', title='Bar Plot: Sales Per Year')
```

---

## 3. **Scatter Plot**

Scatter plot digunakan untuk melihat hubungan antara dua variabel.

### Contoh:
```python
# DataFrame dengan dua kolom numerik
df['Profit'] = [20, 25, 30, 40, 50]

# Membuat Scatter Plot
df.plot(x='Sales', y='Profit', kind='scatter', title='Scatter Plot: Sales vs Profit')
```

---

## 4. **Box Plot**

Box plot digunakan untuk menggambarkan distribusi data, outlier, dan rentang kuartil.

### Contoh:
```python
# Membuat Box Plot
df[['Sales', 'Profit']].plot(kind='box', title='Box Plot: Sales and Profit')
```

---

## 5. **Histogram**

Histogram digunakan untuk menggambarkan distribusi frekuensi data.

### Contoh:
```python
# Membuat Histogram
df['Sales'].plot(kind='hist', bins=5, title='Histogram: Sales Distribution')
```
